**Converting Height and Weight**

```python
df['Weight'] = df['Weight']/2.205
df['Height'] = df['Height'] * 2.54
```

```python
df.rename(columns={'Weight': 'Weight (cms)' , 'Height': 'Height (cms)'} , inplace = True)
```

**Dropping Columns**

```python
df = df.drop(columns='Density')
```

```python
df
```

**Numeric Matric Creation**

```python
X = df.drop(columns=["BodyFat" , "Knee" , "Ankle" , "Biceps" , "Forearm" , "Wrist"])
y = df["BodyFat"]
```

```python
print(X)
```

```python
print(y)
```

No missing values and no categorical variable so no need to handle this all .... 

**Splitting dataset into training and test set**

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size= 0.2 , random_state= 1)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Training Multiple Regression Model on training data**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train , y_train)
```

**Predicting the test set results**

```python
y_pred = regressor.predict(X_test)
```

**Evaluating Model**

*MAE*

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test , y_pred)
```

*R2 Score*

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

**Training Random Forest Regressor on training data**

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=100 , random_state=0)
reg.fit(X_train , y_train)
```

**Predicting Results**

```python
y_pred = reg.predict(X_test)
```

**Evaluating Results**

*MAE*

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test , y_pred)
```

*r2_score*

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

linear model performed better than the random forest maybe the size of dataset is small and body fat linealry depends upon abdomen , hips etc ..
